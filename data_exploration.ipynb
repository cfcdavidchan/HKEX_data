```python
dt = pd.read_csv('./data/MC30_AllFB_20180201.csv', names=full_order_book_header, index_col=False)
dt['date'] = pd.to_datetime(dt['date'], format='%Y%m%d%H%M%S%f')
dt['trade_time'] =  pd.to_datetime(dt["trade_time"].apply(lambda x: format(x, 'f')), format='%Y%m%d%H%M%S%f.000000')
dt.set_index('date', inplace=True)
```

```python
lunch_hour_list = list(dt.between_time('12:00','13:00').index)
```

```python
dt_clean = dt.drop(lunch_hour_list)
```
